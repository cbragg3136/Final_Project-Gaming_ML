```python
y = df["maxprice"]
X = df.drop(columns=["maxprice"])
# X = df[["maxccu", "totaltags", "followers", "positive", "negative"]]
print(X)
```

```python
# y=y.reshape(-1,1)
# y.shape

y= y.values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)

X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
# X_test.fillna(X_test.mean())
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train_scaled, y_train_scaled)
```

```python
score = regressor.score(X_train_scaled, y_train_scaled)
print(score)
```

```python
from sklearn.metrics import mean_squared_error

predictions = regressor.predict(X_train_scaled)
MSE = mean_squared_error(y_train_scaled, predictions)
r2 = regressor.score(X_train_scaled, y_train_scaled)

print(f"MSE: {MSE}, R2: {r2}")
```
