```python
# Dependencies and Setup

import pandas as pd
import numpy as np
```

```python
steam_final_cleaned = pd.read_csv('../Scraped Data/steam_final_cleaned.csv', encoding='utf-8-sig')
steam_final = steam_final_cleaned[['appid', 'metascore', 'recommendations', 'positive', 'negative', 'developer', 'publisher']]
steam_final
```

```python
all_param_binary_tags = pd.read_csv('../reduced_data/all_param_binary_tags.csv')
all_param_binary_tags
```

```python
steam_all_param_binary = pd.merge(steam_final, all_param_binary_tags, on="appid", how="right")
steam_all_param_binary
```

```python
steam_all_param_binary.to_csv('../reduced_data/steam_all_param_binary_tags_merged.csv', encoding='utf-8-sig', index = False)
```

```python
all_param = pd.read_csv('../reduced_data/all_param.csv')
all_param
```

```python
steam_all_param = pd.merge(steam_final, all_param, on="appid", how="right")
steam_all_param
```

```python
steam_all_param.to_csv('../reduced_data/steam_all_param_merged.csv', encoding='utf-8-sig', index = False)
```
